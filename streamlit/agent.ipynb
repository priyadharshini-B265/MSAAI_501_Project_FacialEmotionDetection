```python
import os
# Load the API key from an environment variable or use a default value

GOOGLE_API_KEY = os.getenv('GOOGLE_API_KEY')

print("Using Gemini API Key:", GOOGLE_API_KEY)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import initialize_agent, AgentType,Tool
```

```python
def generate_emotion_summary_agent():

    # Tool function must accept an argument
    def emotion_data(_input: str = None):
        return (
            "Emotion table:\n\n\n"
            "Counts: happy: 5, sad: 3, neutral: 2, angry: 1\n\n\n"
            "Distribution: happy: 50%, sad: 30%, neutral: 15%, angry: 5%"
        )

    tools = [
        Tool(
            name="EmotionData",
            func=emotion_data,
            description="Returns detailed emotion detection statistics."
        )
    ]

    # Gemini LLM
    llm = ChatGoogleGenerativeAI(
        model="gemini-2.5-flash",
        temperature=0.4,
        google_api_key=GOOGLE_API_KEY
    )

    agent = initialize_agent(
        tools,
        llm,
        agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
        verbose=False,
    )

    prompt = """
Analyze the emotion detection results using the EmotionData tool.
Generate a short, clear summary including:
- Number of people detected  
- Most common emotion  
- Emotion distribution  
- Any patterns or insights  
Write professionally and clearly.
"""

    summary = agent.run(prompt)
    return summary
```

```python
agent_summary = generate_emotion_summary_agent()
print("Generated Emotion Summary:", agent_summary)
```
